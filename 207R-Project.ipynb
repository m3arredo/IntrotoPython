##### 2.Analyze the relationship between family size and happiness score across the countries.

We will be using the information provided in the dataset of the 157 countries to understand the relationship between family size and happiness score. We will continue to analyze the top countries that show the most extreme cases.

    1. Choice of Visualization- A scatterplot and correlation will be used to see if there is a postive, negative or no relationship between family and happiness in the countries of the dataset. The correlation heatmap was added to understand the strength of their correlation. We then chose to show the countries with the most extreme cases in the map to determine if the pattern is seen based on regions. 

 2.2 Plot the Graph

```python
sns.scatterplot(data=df, x='Happiness Score', y='Family').set(title='Happiness Score and Family')
df_corr1 = df[['Family','Happiness Score']]
plt.figure(figsize=(10,8))
plt.title('Correlation Heatmap')
sns.heatmap(df_corr1.corr(), annot=True, cmap='Blues', fmt= '.4', linewidth=0.1)
plt.show()
```

```python
#obtaining the top and bottom cases for the relationship
df_top = df[(df['Happiness Score'] > 7) & (df['Family'] > 1.1) ];
df_bottom= df[(df['Happiness Score'] < 4) & (df['Family'] < 0.4)];

import geopandas
import folium 
# Read the geopandas dataset
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
# Merge the two DataFrames together
table = world.merge(df_top, how="right", left_on=['name'], right_on=['Country'])
# Clean data: remove rows with no data
table = table.dropna(subset=['Country'])

# Create a map
my_map = folium.Map()
# Add the data
folium.Choropleth(geo_data=table,name='choropleth',data=table, columns=['Country','Happiness Score'], key_on='feature.properties.name',
                  fill_color='OrRd', fill_opacity=0.7, line_opacity=0.2,legend_name='Happiness Score').add_to(my_map)
my_map
```

```python
# Read the geopandas dataset
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
# Merge the two DataFrames together
table = world.merge(df_bottom, how="left", left_on=['name'], right_on=['Country'])
# Clean data: remove rows with no data
table = table.dropna(subset=['Country'])

# Create a map
my_map = folium.Map()
# Add the data
folium.Choropleth(geo_data=table,name='choropleth',data=table, columns=['Country','Happiness Score'], key_on = 'feature.properties.name',
                  fill_color='OrRd', fill_opacity=0.7, line_opacity=0.2,legend_name='Happiness Score').add_to(my_map)
my_map
```

2.3 Inference : The scatterplot and heatmap show a positive strong correlation. The same values of the previous heatmap for strength level will be considered. Since their correlation is 0.74, we say that their correlation is strong. Their relationship is also seen in the scatterplot. A positive relationship is seen which means that a higher family size shows a higher happiness score. We can say that most countries with low family sizes are less happier compared to those who have more family. We also notice that most countires have at least 0.6 for the varaible 'family'. 

We then analyze the top and bottom cases to see if a pattern can be identified in the map. The top cases are located in two regions- Western Europe, and Australia and New Zealand. The bottom cases are usually located in Sub-Saharan Africa. We could say that we identified a trend in the top and bottom cases. The region in which they live can influence their family size and therefore, the happiness score. However, we would need to continue analysis to confirm our theory from this correlation. 

##### 3. Analyze the difference in economies throughout the world. 

We will be using the column 'Economy" of the 157 counties given in the dataset to see their geographical placement, and if a pattern can be identified for economy.

3.1. Choice of Visualization - We will use a histogram to understand the distribution of economy across the continents. A scatterplot will be then used to see the GDP per Capita for the countries with the highest economy and the lowest. In this case, we are not looking for a relationship in the last two plots, we will look at the points individually. 

  

3.2 Plot Graph

```python
plt.scatter(df['Region'],df['Economy (GDP per Capita)'])
plt.xticks(rotation='vertical')
plt.title('Economy in Continents')
plt.ylabel('GMP per Capita')
```

```python
df_filt = df[df['Economy (GDP per Capita)'] > 1.5]
df_filt2 = df[df['Economy (GDP per Capita)'] < 0.25]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].scatter(df_filt['Country'],df_filt['Economy (GDP per Capita)'])
axs[1].scatter(df_filt2['Country'],df_filt2['Economy (GDP per Capita)'])

#ylabel
axs[0].set_ylabel('Trust (Government Corruption)')
axs[1].set_ylabel('Economy (GDP per Capita)')
#title
axs[0].set_title('Top Countries in Economy ')
axs[1].set_title('Bottom Countries in Economy')

axs[0].set_xticklabels(df_filt['Country'], rotation=90);
axs[1].set_xticklabels(df_filt2['Country'], rotation=90);
```

3.3 Inference: From the first graph, we can determine the difference in Economy (GDP per Capita) from each Region. Some regions have a closer distribution in Economy such as Western Europe, but other regions have a more seperate distribution such as Sub-Saharan Africa. This makes sense since the countries in Sub-Saharan Africa can be very different in qualities including Economy. The top countries include Qatar, Luxembourg and Singapore in Economy. The bottom countries are Somalia, Congo, and Burundi. The top countries do not seem to follow a trend, but all the lower countries are from the Sub-Saharan Africa. 

We can identify from the top graph, and the bottom right that the Sub-haran Africa is significantly lower in economy than other regions. This can help in focusing programs of need to this area.
